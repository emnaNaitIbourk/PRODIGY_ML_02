```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
data= pd.read_csv("Mall_Customers.csv")
data.head()
data.columns
```

```python
numeric_columns=data.select_dtypes(include=[np.number]).columns
non_numeric_columns=data.select_dtypes(exclude=[np.number]).columns
```

```python
data[numeric_columns]=data[numeric_columns].fillna(data[numeric_columns].mean())
```

```python
label_encoder=LabelEncoder()
for column in non_numeric_columns:
 data[column]=label_encoder.fit_transform(data[column].astype(str))
```

```python
features=['Age', 'Annual Income (k$)','Spending Score (1-100)']
X=data[features]
#scale numerical features
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
#optimal number of clusters using Elbow method 
sse=[]
for k in range (1,11):
    kmeans=KMeans(n_clusters=k,random_state=42)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,11),sse,marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
plt.title("Elbow Method")
plt.show()
```

```python
silhouette_scores=[]
for k in range(2,11):
    kmeans=KMeans(n_clusters=k,random_state=42)
    kmeans.fit(X_scaled)
    silhouette_scores.append(silhouette_score(X_scaled,kmeans.labels_))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(2,11),silhouette_scores,marker="o")
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Scores")
plt.title("Silhouette Score Method")
plt.show()
```

```python
optimal_k=5
kmeans=KMeans(n_clusters=optimal_k,random_state=42)
data['Cluster']=kmeans.fit_predict(X_scaled)
```

```python

colors=['red','blue','purple','green','yellow']
plt.figure(figsize=(10,6))
sns.scatterplot(data=data,x='Annual Income (k$)',y='Spending Score (1-100)',hue='Cluster',palette=colors)
plt.title("Clusters of Customers")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.legend()
plt.show()
```

```python
cluster_centers=scaler.inverse_transform(kmeans.cluster_centers_)
cluster_centers_df=pd.DataFrame(cluster_centers,columns=features)
cluster_centers_df['Cluster']=range(optimal_k)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(cluster_centers_df.set_index('Cluster'), annot=True,cmap='coolwarm')
plt.title("Cluster Centers")
plt.show()
```

```python
sns.pairplot(data,hue='Cluster',palette='viridis',vars=features)
plt.show()
```

```python
plt.figure(figsize=(15,10))
for i,feature in enumerate(features):
    plt.subplot(2,2, i+1)
    sns.boxplot(x='Cluster',y=feature,data=data,palette='viridis')
    plt.title(f'Distribution of {feature} in Clusters')
    plt.tight_layout()
    plt.show()
```

```python
cluster_summary=data.groupby('Cluster')[features].mean().reset_index()
print(cluster_summary)
#save the clustered data
data.tocsv('clustered_customers.csv',index=False)
```
